```python
from sklearn import datasets
import numpy as np
import pandas as pd
```

```python
iris=datasets.load_iris()
x=iris.data
y=(iris.target!=0)*1
#print(x)
y = y.reshape((y.shape[0],1))
```

```python
x = np.append(x,np.ones((x.shape[0],1)),axis=1)
x=x[:,:5]
x
```

```python
from sklearn import model_selection
```

```python
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
def sigmoid(x,m):
    ans = 1/(1.0+np.exp(-(m*x).sum()))
    #print("sigm",ans)
    return ans


def sgd(x_train,y_train,learning_rate,m):
    m_slope=np.zeros(x_train.shape[1])
    M=len(x_train)
    for i in range(M):
    
        x=x_train[i]
        y=y_train[i]
        ans=0;
        for k in range(x_train.shape[1]):
            ans+=sigmoid(m[k],x[k])
        #print(ans)
        for j in range(x_train.shape[1]):
            m_slope[j]+=(1/M)*(y-ans)*x[j]
        
    new_m=m-learning_rate*m_slope
    return new_m
def gd(x_train,y_train, learning_rate, num_iterations):
    m=np.zeros(x_train.shape[1])

    for i in range(num_iterations):
        m= sgd(x_train,y_train, learning_rate, m)
        print(i, " Cost: ", cost(x_train,y_train, m))
    return m
def cost(x_train,y_train, m):
    total_cost = 0
    M = len(x_train)
    for i in range(M):
        x = x_train[i]
        y = y_train[i]
        total_cost += (1/M)*(-y * np.log(sigmoid(x,m))) - (1 - y) * np.log(1 - sigmoid(x,m))
    return total_cost

    
    
    
```

```python
def run():
    x_train=x
    y_train=y
    learning_rate = 0.000001
    num_iterations = 60
    m = gd(x_train,y_train, learning_rate, num_iterations)
    return m
```

```python
y_pred=np.zeros(len(y_test))
print(y_pred)
m=run()
print(m)
print(x_test[1])
for i in range(len(x_test)):
    y_pred[i]=(x_test[i]*m).sum()
print(y_pred)
for i in range(len(y_pred)):
    if(y_pred[i]>0.005):
        y_pred[i]=1
    else:
        y_pred[i]=0
print(y_pred)
np.savetxt("predicted_data.csv", y_pred,delimiter = ",",fmt='%1.5f')
#print(y_test)
z=np.zeros(len(y_test))
z=y_pred-y_test
print(z)
```
